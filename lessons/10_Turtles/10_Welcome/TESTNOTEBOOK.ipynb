## **Let's Test it Out!**

Now that we have ipyturtle3 installed, we can test it out by creating a simple turtle graphics program. In the next cell, let's import the necessary libraries and create a turtle object.

```python
# Run Me!

import ipyturtle3 as turtle
from ipyturtle3 import hold_canvas, Canvas, TurtleScreen, Turtle
from IPython.display import display

# Create a Canvas
myCanvas = Canvas(width=800, height=250)
display(myCanvas)

# Create a TurtleScreen
myTS = TurtleScreen(myCanvas)
myTS.clear()

# Create a Turtle
eric = Turtle(myTS)
eric.shape("turtle")
eric.color("green")
eric.speed(5)

# Functions to draw each letter
def draw_L(t):
    t.setheading(-90)
    t.forward(125)
    t.setheading(0)
    t.forward(50)

def draw_E(t):
    t.backward(50)
    t.setheading(-90)
    t.forward(60)
    t.setheading(0)
    t.forward(50)
    t.backward(50)
    t.setheading(-90)
    t.forward(65)
    t.setheading(0)
    t.forward(50)

def draw_A(t):
    t.setheading(-75)
    t.forward(130)
    t.backward(130)
    t.setheading(75)
    t.backward(130)
    t.forward(62.5)
    t.setheading(0)
    t.forward(35)

def draw_G(t):
    t.backward(50)
    t.setheading(-90)
    t.forward(125)
    t.setheading(0)
    t.forward(50)
    t.setheading(90)
    t.forward(62.5)
    t.setheading(180)
    t.forward(25)
    t.setheading(0)

def draw_U(t):
    t.setheading(-90)
    t.forward(125)
    t.setheading(0)
    t.forward(50)
    t.setheading(90)
    t.forward(125)
    t.setheading(0)

# Positions and corresponding drawing functions to reproduce 'LEAGUE'
letters = [
    (-350, 75, draw_L),
    (-225, 75, draw_E),
    (-160, 75, draw_A),
    (-50, 75, draw_G),
    (-25, 75, draw_U),
    (100, 75, draw_E),
]

# Loop through positions and draw each letter (pen up to move, pen down to draw)
for x, y, fn in letters:
    eric.penup()
    eric.goto(x, y)
    eric.pendown()
    fn(eric)
```
