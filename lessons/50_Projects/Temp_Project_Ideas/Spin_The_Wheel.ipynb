Typically, a *Spin The Wheel* game consists of a wheel divided into sections, each representing a different prize. Players take turns spinning the wheel, and wherever the wheel stops determines the prize they win. However, feel free to add your own twist to the game! This is your chance to get creative and have fun.

Software Development is all about problem-solving and creativity. So, think about how you can make your game engaging and enjoyable for players.

It also might be a pretty good time to introduce the concept of the Software Development Life Cycle (SDLC), let's take a look. 

<details style="cursor: pointer;"><summary style="font-size:1.5em;font-weight:700;margin:0;">The Software Development Life Cycle (SDLC)</summary>

The SDLC is a structured approach to software development that consists of several phases, including planning, design, implementation, testing, and deployment. It's a helpful framework to ensure that software is developed efficiently and meets the desired requirements.

- **Planning:** This is where you would think about what features you want in your game, like the number of sections on the wheel, or maybe types of prizes, or even how players will interact with it on a logical level. Is the prize monetary, points-based, or something else? Maybe it's just some funny message?
<br>

- **Design:** Sketch out how the game will flow, including user inputs and outputs.
<br>
- **Implementation:** Write the code to create the game using a programming language of your choice (e.g., Python, JavaScript).
<br>
- **Testing:** Play the game multiple times to ensure it works as expected and fix any bugs.
<br>
- **Deployment:** Share your game with friends or online for others to play.
</details>
