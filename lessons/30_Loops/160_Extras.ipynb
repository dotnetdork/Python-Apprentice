### **Euler's FizzBuzz**

This version relies on <span title="Arithmetic where numbers wrap around after hitting a set limit (the modulus)." style="cursor: help;"><strong>modular arithmetic</strong><svg style="width:18px;height:18px; vertical-align: middle; margin-left: 2px; margin-bottom: 3px;" viewBox="0 0 24 24"><path fill="currentColor" d="M12,2A10,10 0 0,0 2,12A10,10 0 0,0 12,22A10,10 0 0,0 22,12A10,10 0 0,0 12,2M12,4A8,8 0 0,1 20,12A8,8 0 0,1 12,20A8,8 0 0,1 4,12A8,8 0 0,1 12,4M11,16.5V11.5H13V16.5H11M11,9.5V7.5H13V9.5H11Z"/></svg></span> and uses $n^4$ $\%$ $15$ to choose what to print.

The remainder then feeds a small lookup table as shown below:

- `1` â†’ the number itself
- `6` â†’ `Fizz`
- `10` â†’ `Buzz`
- `0` â†’ `FizzBuzz`

```python
# Run Me!

# Euler-style mapping from remainder (n^4 % 15) to the correct label
fb = [(lambda n: {1: n, 6: "Fizz", 10: "Buzz", 0: "FizzBuzz"}[n ** 4 % 15])(n) for n in range(1, 101)]
print(fb)
```
