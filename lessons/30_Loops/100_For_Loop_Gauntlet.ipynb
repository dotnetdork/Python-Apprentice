## **Nested For-Loops**

Now let's practice writing loops inside of loops! These are called *nested* `for`*-loops*. 

````python
# Examples of Nested Loops

for i in range(8):
    for j in range(8):
        print(f"({i},{j})", end= ' ')
    print()
````
